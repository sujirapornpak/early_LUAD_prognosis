### STEP 7.2: Remove sum row = 0

```python
df72 = df71.copy()
```

```python
# How to remove all (columns and) rows that sum to 0
# https://stackoverflow.com/questions/23573052/pandas-dataframe-how-do-i-remove-all-columns-and-rows-that-sum-to-0
df72 = df72.loc[(df72.sum(axis=1) != 0),:]
with pd.option_context('display.max_rows',2): display(df72) ## 505 - 470 = 35 samples were removed
```

```python
dff72 = dff71.copy()
# How to remove all (columns and) rows that sum to 0
# https://stackoverflow.com/questions/23573052/pandas-dataframe-how-do-i-remove-all-columns-and-rows-that-sum-to-0
dff72 = dff72.loc[(dff72.sum(axis=1) != 0),:]
with pd.option_context('display.max_rows',2): display(dff72) ## 477 samples same because 
# make pivot table after drop genes (while selecting 47 driver genes)
```
