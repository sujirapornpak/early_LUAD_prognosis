#### --> For X1,y1 (featu107_ridge) 

```python
#example 12-17: Searching hyperparameter by GridSearchCV with continuing multiple models

from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
import time


time_start = time.time()

#----- KNN -----
knn = KNeighborsClassifier()
knn_params = {'n_neighbors': range(1, 50, 2)} #1, 3, 5,...

knn_grid = GridSearchCV(estimator=knn, param_grid=knn_params, cv=5) #cv: default=5 (using StratifiedKFold if cv=None/integer)
knn_grid.fit(X1_train,y1_train)

print('KNN Best Score:', knn_grid.best_score_) # Result: 0.6044326241134753
print('KNN Best Params:', knn_grid.best_params_) # Result: 1
print()

#----- Random Forest -----
forest = RandomForestClassifier(random_state=42)
forest_params = {
    'n_estimators': range(10, 100, 5), # n_estimators=The number of trees in the forest.
    'criterion': ['entropy', 'gini'],
    #'max_depth': range(1, 80)
}
forest_grid = GridSearchCV(estimator=forest, param_grid=forest_params, cv=5)
forest_grid.fit(X1_train,y1_train)

print ('Random Forest Best Score:', forest_grid.best_score_) # Result: 0.8152482269503546
print ('Random Forest Best Params:', forest_grid.best_params_) # Result: {'criterion': 'entropy', 'n_estimators': 40}
print()

#----- SVM -----
# original from book= 'C': [0.1, 1, 5, 10, 50], 'gamma': range(1, 10)

svm = SVC()
svm_params = [
    {'kernel': ['linear'], 'C': [0.1, 1, 5, 10, 50]},
    {'kernel': ['rbf'], 'C': [0.1, 1, 5, 10, 50], 'gamma': [0.0001,0.001,0.01,0.1,1,2,4,6,8,10]},
    {'kernel': ['poly'], 'C': [0.1, 1, 5, 10, 50], 'degree': [0.0001,0.001,0.01,0.1,1,2,4,6,8,10]},
]

svm_grid = GridSearchCV(estimator=svm, param_grid=svm_params, cv=5)
svm_grid.fit(X1_train,y1_train)

print('SVM Best Score:', svm_grid.best_score_) # Result: 0.8572695035460993
print('SVM Best Params:', svm_grid.best_params_) # Result: {'C': 50, 'degree': 1, 'kernel': 'poly'}
print()

#----- Multi-layer Perceptron -----
mlp = MLPClassifier(activation='relu',
                   solver='lbfgs', #change from lbfgs (default=adam, which is suitable for big data (>1000 samples)
                   max_iter=200000,   #increase interations from 10000
                   random_state=42)
mlp_params = { 'hidden_layer_sizes': range(3, 10) }   #ทดสอบ 3 - 10 นิวรอน

mlp_grid = GridSearchCV(estimator=mlp, param_grid=mlp_params, cv=5)
mlp_grid.fit(X1_train,y1_train)

print('MLP Best Score:', mlp_grid.best_score_) # Result: 0.8781914893617021 (solver='lbfgs')
print('MLP Best Params:', mlp_grid.best_params_) # Result: {'hidden_layer_sizes': 8} (solver='lbfgs')
print()

#Note: The default solver ‘adam’ works pretty well on relatively large datasets (with thousands of 
# training samples or more) in terms of both training time and validation score. 
# For small datasets, however, ‘lbfgs’ can converge faster and perform better.

# mlp_best = mlp_grid.best_estimator_

#---------------------------------------

time_end = time.time()
print(f'Total Time: {time_end - time_start} - Second')
```

```python
## Model building with the determined parameters
clf1 = RandomForestClassifier(n_estimators=40, criterion='entropy',random_state=42)
clf2 = SVC(kernel='poly',C=50,degree=1)
clf3 = MLPClassifier(activation='relu',
                   solver='lbfgs', #change from lbfgs (default=adam, which is suitable for big data (>1000 samples)
                   max_iter=200000,   #increase interations from 10000
                   hidden_layer_sizes=(8), # (8)=1 layer and 8 neurons; DON'T use (4,4) = 2 layers, each layer = 4 neurons
                   random_state=42)
```

```python
## Model training 
clf1.fit(X1_train,y1_train)
clf2.fit(X1_train,y1_train)
clf3.fit(X1_train,y1_train)
```

```python
#------Model Evaluation and Confusion matrix------

## https://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html
## https://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html

from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
y1_predRF = clf1.predict(X1_test)
y1_predSVC = clf2.predict(X1_test)
y1_predMLP = clf3.predict(X1_test)
```

```python
display(confusion_matrix(y1_test,y1_predRF), 
       confusion_matrix(y1_test,y1_predSVC),
        confusion_matrix(y1_test,y1_predMLP))
# Row sum: y1_test
# clus1 = 23 (e.g. SVC: 22+0+1)
# clus2 = 40 (e.g. SVC: 2+38+0)
# clus3 = 17 (e.g. SVC: 2+1+14)
```

```python
display(accuracy_score(y1_test,y1_predRF, normalize=False), #TP 17 + TN 37 + TN 14 = 68
       accuracy_score(y1_test,y1_predSVC, normalize=False),
       accuracy_score(y1_test,y1_predMLP, normalize=False)) 
```

```python
class_names = ['clus1', 'clus2','clus3']
print('RF for X1,y1')
print(classification_report(y1_test,y1_predRF, target_names=class_names))
print('SVC for X1,y1')
print(classification_report(y1_test,y1_predSVC, target_names=class_names))
print('MLP for X1,y1')
print(classification_report(y1_test,y1_predMLP, target_names=class_names))
```

```python
from collections import Counter
from matplotlib_venn import venn2, venn3, venn3_circles,venn2_circles   #import matplotlib_venn as venn 
import matplotlib.pyplot as plt

# $ conda install -c conda-forge matplotlib-venn=0.11.5 (for py3.6 env)
# Ref: https://towardsdatascience.com/how-to-create-and-beautify-venn-diagrams-in-python-331129bd4ed3
```

```python

# Ref:https://towardsdatascience.com/how-to-create-and-beautify-venn-diagrams-in-python-331129bd4ed3

vd2=venn2([A1,A2],
          set_labels=('Ridge_coef_107genes(c1,c2,c3)', 'Ridge_coef_98genes(c1,c3)'),
          set_colors=('#c4e6ff', '#F4ACB7'), alpha = 0.8) #'#9D8189'
venn2_circles([set(A1),set(A2)], linestyle='-', linewidth=1, color='grey') # linestyle='-.'
for text in vd2.set_labels:
    text.set_fontsize(16);
for text in vd2.subset_labels:
    text.set_fontsize(16)
plt.title('Venn Diagram for 2 feature selection methods',fontname='Times New Roman',fontweight='bold',fontsize=20,
 pad=30,backgroundcolor='#cbe7e3',color='black',style='italic');
plt.show()
```
