## STEP 1: Prepare dataframe of heat diffusion score of 2298 nodes (genes) by combining clus1 (90), clus2(158) and clus3(70)

### STEP 1.1: Open previous 3 tsv files from Cytoscape_Heatdiffusion folder

```python
# Open heat score output files:

# cluster1 (90 samples)
df11=pd.read_csv("./output/1.1_OUT_HeatDiffScore_2298genes_sortAvg_clus1-90_12Apr.tsv",sep='\t',header=0)
# cluster2 (158 samples)
df12=pd.read_csv("./output/1.1_OUT_HeatDiffScore_2298genes_sortAvg_clus2-158_12Apr.tsv",sep='\t',header=0)
# cluster3 (70 samples)
df13=pd.read_csv("./output/1.1_OUT_HeatDiffScore_2298genes_sortAvg_clus3-70_12Apr.tsv",sep='\t',header=0)

df11.head(2)
```

```python
# Drop average column
df11=df11.drop(['avg_clus1'], axis=1)
df12=df12.drop(['avg_clus2'], axis=1)
df13=df13.drop(['avg_clus3'], axis=1)

# Sort gene name (from A-Z)
df11=df11.sort_values(by='gene', ascending=True)
df12=df12.sort_values(by='gene', ascending=True)
df13=df13.sort_values(by='gene', ascending=True)

# after sorting, then reset index
df11=df11.reset_index(drop=True)
df12=df12.reset_index(drop=True)
df13=df13.reset_index(drop=True)

# Make 'gene' column into index because it's required before transpose
df11g = df11.set_index('gene')
df12g = df12.set_index('gene')
df13g = df13.set_index('gene')

with pd.option_context('display.max_rows',4): display(df11g)
```

### STEP 1.2: Transpose columns (samples) into rows and genes to columns (=features)

```python
# transpose columns into rows ** Note: The first 'gene' column must be index so that it will be col names.
df11TP=df11g.transpose() # 90 rows × 2298 columns
df12TP=df12g.transpose() # 158 rows × 2298 columns
df13TP=df13g.transpose() # 70 rows × 2298 columns

# Tip 1.2: Change index name (cannot use .rename() method for index)
df11TP.index.names = ['sample_id']
df12TP.index.names = ['sample_id']
df13TP.index.names = ['sample_id']

# Turn index (sample_id) back to column to prepare for merging the cluster names 
df11tps = df11TP.reset_index()
df12tps = df12TP.reset_index()
df13tps = df13TP.reset_index()
df11tps.head(3)

with pd.option_context('display.max_rows',2): display(df11tps)
```

### STEP1.3: Open file with specified cluster and sample_id for merging clusters (c1,c2,c3) as target (Y) 

```python
dfclus=pd.read_csv("8.1_OUT_cSIF_Driv47_pivot_merge3clusterpyNBS_318samp_22Mar.tsv",sep='\t',header=0)

#Merge 3 clusters assigned for 318 samples (inner merge)
df11clus = df11tps.merge(dfclus[['sample_id', 'cluster']],on='sample_id',how='inner') 
df12clus = df12tps.merge(dfclus[['sample_id', 'cluster']],on='sample_id',how='inner') 
df13clus = df13tps.merge(dfclus[['sample_id', 'cluster']],on='sample_id',how='inner') 

with pd.option_context('display.max_rows',2): display(df11clus)
```

```python
# Concat 3 dataframes
# To join dataframe by .concat() --> https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html

df14 = pd.concat([df11clus, df12clus, df13clus],ignore_index=True) # default: ignore_index=False

with pd.option_context('display.max_rows',2): display(df14)
```

```python
print(df14.cluster.unique().tolist())
# Query string
display(len(df14.query("cluster == 'c1'")),
       len(df14.query("cluster == 'c2'")),
        len(df14.query("cluster == 'c3'")))
```

```python
# Export/convert dataframe to csv or tsv file
df14.to_csv('./output/1.1_LUAD_ML_OUT_2298Featu318Sam_3class_14Jul22.tsv', sep="\t",index=False) 
```

### STEP2.1: Prepare df

Useful link of feature selection:

```python
# Open file prepared for ML:
df21 = pd.read_csv("./output/1.1_LUAD_ML_OUT_2298Featu318Sam_3class_14Jul22.tsv",sep='\t',header=0)
with pd.option_context('display.max_rows',2): display(df21)
```

```python
# Encoding class c1,c2,c3
from sklearn.preprocessing import LabelEncoder

df22 = df21.copy()

# fit() followed by .transform()
encoder = LabelEncoder()
encoder.fit(df22['cluster']) # Don't apply .fit() twice, next encoding can use .transform() directly
df22['cluster']=encoder.transform(df22['cluster']) 

# Try to decode back from 0,1,2 into string
inv= encoder.inverse_transform([0, 1, 2])
print(inv)
```

```python
## Define x (covariates or independent variables (features)) and y (response or dependent variables (classes)) 
x=df22.drop(['sample_id','cluster'],axis=1) #drop column 'sample_id' because use index no. instead
y=df22['cluster']
```

### **STEP2.2: Apply Train-Test split BEFORE Feature Selection!!!

Ref1: https://stackoverflow.com/questions/56308116/should-feature-selection-be-done-before-train-test-split-or-after

Ref2: https://www.nodalpoint.com/not-perform-feature-selection/

**Main Reason:** 

The conventional answer is to do feature selection after splitting as there can be **INFORMATION LEAKAGE**, if done before, from the Test-Set.

When having such doubts, it is useful to imagine that you simply do not have any access in any test set during the model fitting process (which includes feature importance); you should treat the test set as literally unseen data (and, since unseen, they could not have been used for feature importance scores).

Question: the trends in the Test Set may be different??
ANS: A standard (but often implicit) assumption here is that the training & test sets are qualitatively similar; it is exactly due to this assumption that we feel OK to just use simple random splits to get them. If we have reasons to believe that our data change in significant ways (not only between train & test, but during model deployment, too), the whole rationale breaks down, and completely different approaches are required.

**Wrap up: the test set is there solely for performance assessment of your model, and it should not be used in any stage of model building, including feature selection.**

```python
## Split the data into 75 to 25 with stratified between y class labels (get approx equal proportion per class)
from sklearn.model_selection import train_test_split
X1_train,X1_test,y1_train,y1_test=train_test_split(x,y,stratify=y,test_size=0.25,random_state=42)

print(X1_train.shape)   ## 0.75*318 = 238.5 samples
print(y1_train.shape)  ## attributes (features) = 2298
print(X1_test.shape)
print(y1_test.shape)

## View the counts of case and control sample in the training set
display(pd.crosstab(y1_train,columns=0),  
        pd.crosstab(y1_test,columns=0))
```

### STEP2.3.1: Try Feature selection using SelectKBest, chi2 to get "20 best features"

```python
from sklearn.feature_selection import SelectKBest, chi2
```

```python
# Ref1: https://towardsdatascience.com/5-feature-selection-method-from-scikit-learn-you-should-know-ed4d116e4172
# Ref2: https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.SelectKBest.html#sklearn.feature_selection.SelectKBest

#Select top 120 features based on chi-square
selector1 = SelectKBest(chi2, k=120)
selector1.fit(X1_train,y1_train) # Only apply .fit() to observe selected features, not yet transform
genechi120 = np.array(X1_train.columns[selector1.get_support()])
len(genechi120)
```

### STEP2.3.2: Try Feature selection using SelectFromModel
#### 1) Using LogisticRegression
Ref:https://towardsdatascience.com/5-feature-selection-method-from-scikit-learn-you-should-know-ed4d116e4172

Note: SelectFromModel feature selection is based on the importance attribute (often is coef_ or feature_importances_ but it could be any callable) threshold. By default, the threshold is the mean.

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LogisticRegression
```

```python
# #Selecting the Best important features according to Logistic Regression using SelectFromModel
# By default, the threshold is the mean

sfm_selector1 = SelectFromModel(estimator=LogisticRegression())
sfm_selector1.fit(X1_train,y1_train)
X1_train.columns[sfm_selector1.get_support()]
```

```python
genelr334 = np.array(X1_train.columns[sfm_selector1.get_support()])
len(genelr334)
```

#### 2) Model-based and sequential feature selection 
Ref: https://scikit-learn.org/stable/auto_examples/feature_selection/plot_select_from_model_diabetes.html#sphx-glr-auto-examples-feature-selection-plot-select-from-model-diabetes-py


```python
#---------------Feature importance from coefficients--------------

import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import RidgeCV

# alphasndarray of shape (n_alphas,), default=(0.1, 1.0, 10.0)
alphas=np.logspace(-6, 6, num=5)
alphas
```

```python
ridge = RidgeCV(alphas=np.logspace(-6, 6, num=5)).fit(X1_train,y1_train) # alphas default=(0.1, 1.0, 10.0)
importance = np.abs(ridge.coef_)
feature_names = np.array(X1_train.columns) #(diabetes.feature_names)

# Plot only the first 50 genes
plt.bar(height=importance[0:51], x=feature_names[0:51])
plt.title("Feature importances via coefficients")
plt.show()
```

```python
df23= pd.DataFrame(importance,index=feature_names)
df23=df23.reset_index()
df23.rename(columns={0: 'Important_coef_','index':'Gene'},inplace=True)
df23.head(2) #df23.shape = (2298, 2)

df23['Important_coef_'].describe()
```

```python
#*************** Select the best features <107 (78 genes) and >107 (132 genes) ***************

df25_a = (df23.query("Important_coef_ >200")) # 24 genes
df25_aa = np.array(df25_a['Gene'])

df25_b = (df23.query("100 < Important_coef_ <= 200")) # 26 genes
df25_bb = np.array(df25_b['Gene'])

df25_c3 = (df23.query("50 < Important_coef_ <= 100")) # 57 genes ** Original condition
df25_cc3 = np.array(df25_c3['Gene'])

# Get gene names with scores > 50 ** Original condition
df25_concat3 = np.concatenate((df25_aa,df25_bb,df25_cc3), axis=None)
print(len(df25_concat3)) # = 107 (24+26+57)
```

### STEP2.4 VENN diagram to conclude 3 feature selecting methods 

```python
from collections import Counter
from matplotlib_venn import venn2, venn3, venn3_circles   #import matplotlib_venn as venn 
import matplotlib.pyplot as plt

# $ conda install -c conda-forge matplotlib-venn=0.11.5 (for py3.6 env)
# Ref: https://towardsdatascience.com/how-to-create-and-beautify-venn-diagrams-in-python-331129bd4ed3
```

```python
##-----------------CONCLUSION: Venn diagram of 3 feature selecting methods----------------- 
#1) len(genechi120)--> Methods= SelectKBest, chi2
#2) len(genelr334) --> Methods= SelectFromModel, LogisticRegression()
#3) len(df25_concat3) = 107(24+26+57) --> Methods= Feature importances via ridge's coefficients, RidgeCV()

A1 = set(df25_concat3) # 107 genes whose ridge's coefficients: 50 < Important_coef_ 
B1 = set(genelr334) # 334 genes from Methods= SelectFromModel, LogisticRegression()
C1 = set(genechi120) # 120 genes from Methods= SelectKBest, chi2

print(len(A1),len(B1),len(C1)) # 107 334 120
```

```python
# Ref:https://towardsdatascience.com/how-to-create-and-beautify-venn-diagrams-in-python-331129bd4ed3

vd3=venn3([A1,B1,C1],
          set_labels=('Ridge_coef_107', 'LogisticR_334','Chi2_120'),
          set_colors=('#c4e6ff', '#F4ACB7','#9D8189'), alpha = 0.8)
venn3_circles([set(A1),set(B1),set(C1)], linestyle='-', linewidth=1, color='grey') # linestyle='-.'
for text in vd3.set_labels:
    text.set_fontsize(16);
for text in vd3.subset_labels:
    text.set_fontsize(16)
plt.title('Venn Diagram for 3 feature selection methods',fontname='Times New Roman',fontweight='bold',fontsize=20,
 pad=30,backgroundcolor='#cbe7e3',color='black',style='italic');
plt.show()
```

```python
# Change set to list:
fea1=list(A1)
fea1.sort(reverse=False)

d = {'Features_Ridge': fea1}
df_AA = pd.DataFrame(data=d)
with pd.option_context('display.max_rows',4): display(df_AA)
```
