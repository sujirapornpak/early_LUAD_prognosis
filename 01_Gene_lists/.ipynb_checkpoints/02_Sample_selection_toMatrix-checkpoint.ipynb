# L02_Sample selection based on LUAD-specific driver genes and three benchmark cancer gene panels
### Objectives:
1) From total 505 samples and 16057 genes (Matrix/pivot file), select samples based on mutations available in those gene panels

2) To export five Matrix files for NBS propagation (Lab_03)

3) Those five gene panels are:
  - 1) 47 early LUAD drivers
  - 2) 151 remaining LUAD drivers (but available only 150 genes)
  - 3) FoundationOne_324 (but available only 299 genes)
  - 4) Truseq_48 (but available only 47 genes)
  - 5) PanCan_127 (but available only 121 genes)

```python
##STEP 1.1: Import required python libraries 

%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## STEP 1: Retrieve lists of LUAD driver genes from three resources

```python
df11 = pd.read_csv('SuppTable1_All_FiveDrivers_GeneList.tsv', sep="\t",header=0)
with pd.option_context('display.max_rows',2): display(df11)
```

```python
# Extra Preprocessing: How to merge 3 sources of LUAD drivers to 198 drivers and 
# subtract 47 from 198, resulting in the remaining 151 drivers  

# Step 1: List all driver sets
driv1_181 =df11['Rubio-Perez(2015)_181'].dropna().unique().tolist() # len()=181
driv2_20 =df11['Bailey(2018)_20'].dropna().unique().tolist() # len()=20
driv3_34 =df11['Bailey(2020)_34'].dropna().unique().tolist() # len()=34
driv4_fin47 =df11['LUAD-Early_47'].dropna().unique().tolist() # len()=47

# Step 2: Concat gene list
drive_concat = np.concatenate((driv1_181,driv2_20,driv3_34), axis=None) #len()=235
drive_concat198 = set(drive_concat) # len(drive_concat198) = 198

# Step 3: Subtract 198-47 = 151 drivers
concat198 = list(drive_concat198)
driv151=[]
for i in list(range(len(concat198))):
    if concat198[i] not in driv4_fin47:
        driv151.append(concat198[i])  # len(driv151) = 151
        
# Step 4: Combine into single table 
concat198.sort(reverse=False)
driv151.sort(reverse=False)

## 198 drivers
df_AA = pd.DataFrame(concat198)
df_AA.rename(columns={0:'Pooled_198Drivers'},inplace=True) # column 1

## 151 drivers
df_BB = pd.DataFrame(driv151) 
df_BB.rename(columns={0:'Remaining_151Drivers'},inplace=True) # column 2
df_CC = pd.concat([df_AA, df_BB], axis=1)

# Export pre-processed tsv files
df_CC.to_csv('./output/ListDrivers198&151.tsv', sep="\t",index=False)  
```

## STEP 3: Select samples based on mutations present in gene panels for NBS propagation
- NBS propagation based on 47 early drivers
- NBS propagation based on 151 remaining drivers
- NBS propagation based on three benchmark cancer gene panels:
  - 1) FoundationOne F1CDx_324 --> 299 genes
  - 2) Trueseq 48 --> 47 genes
  - 3) Pancan 127 --> 121 genes

```python
# Open cleaned Matrix (pivot) file:
df31 = pd.read_csv("../01_Somatic_variant_preprocessing/output/6.3_LUAD_snv_mc3_OUT_cleaned_cSIF_pivot_16057genes.tsv",sep='\t',header=0)
with pd.option_context('display.max_rows',2): display(df31) # 505 samples Ã— 16057 genes
```

```python
# Set index (multi-index)
df32 = df31.set_index(['tumor_stage', 'sample_id']) 
dfindex = df32.iloc[:,0:0] # Prepare index
dfindex.head(1)
```

## ***Checking whether preferred genes are available from all 16057 genes

```python
# Before selecting, check whether preferred genes are available
All16057 = list(df32.columns) # Get list of all gene names
len(All16057) # 16057 genes

# 1) Checking for 198 (47 and 151) LUAD drivers
notlistA = []
for i in list(range(len(driv4_fin47))):
    if driv4_fin47[i] not in All16057: # 47 early LUAD
        notlistA.append(driv4_fin47[i])

notlistB = []
for i in list(range(len(driv6_151))):
    if driv6_151[i] not in All16057: # 151 remaining LUAD
        notlistB.append(driv6_151[i])

# 2) Checking for three cancer gene panels        
notlistC = []
for i in list(range(len(driv10_F1CDx299))):
    if driv10_F1CDx299[i] not in All16057: # FoundationOne (N = 299)
        notlistC.append(driv10_F1CDx299[i])

notlistD = []
for i in list(range(len(driv11_Truseq47))):
    if driv11_Truseq47[i] not in All16057: # Trueseq (N = 47)
        notlistD.append(driv11_Truseq47[i])

notlistE = []
for i in list(range(len(driv12_Pancan121))):
    if driv12_Pancan121[i] not in All16057: # PanCan (N = 121)
        notlistE.append(driv12_Pancan121[i])

print("From 47 early LUAD, gene(s) which are not available are",notlistA)
print("From 151 remaining LUAD, gene(s) which are not available are",notlistB)
print("From FoundationOne (N = 299), gene(s) which are not available are",notlistC)
print("Trueseq (N = 47), gene(s) which are not available are",notlistD)
print("PanCan (N = 121), gene(s) which are not available are",notlistE)
```

```python
# Checking for an alias gene names:
'LPHN2' in All16057 # Aliases for ADGRL2 gene are LPHN2, ..., ...
```

```python
# Checking for an alias gene names:
'AGO3' in All16057 # Aliases for EIF2C3 gene are AGO3, ..., ...
```

```python
'HHR23B' in All16057 # Aliases for RAD23B gene are P58, HR23B, HHR23B; but all are unavailable
```

```python
# Thus, modify remaining LUAD drivers from 151 --> 150 genes
## 1) Remove RAD23B gene 
driv6_151.remove('RAD23B')

## 2) Replace 'EIF2C3' with 'AGO3'
driv6_150new = [g.replace('EIF2C3', 'AGO3') for g in driv6_151]

## 3) Replace 'ADGRL2' with 'LPHN2'
driv4_fin47new = [g.replace('ADGRL2', 'LPHN2') for g in driv4_fin47]

len(driv6_150new)
```

### Step 3.1: 477 samples having mutations in 47 early LUAD drivers

```python
# Prepare index
dfMat_LUAD47a = dfindex.copy()
 
# For loop to append columns of 47 driver genes
for i in list(range(len(driv4_fin47new))): 
    dfMat_LUAD47a[driv4_fin47new[i]] = np.array(df32[driv4_fin47new[i]])

# Keep only samples having at lease one mutated genes
dfMat_LUAD47b = dfMat_LUAD47a[dfMat_LUAD47a.sum(axis=1) >= 1]
with pd.option_context('display.max_rows',2): display(dfMat_LUAD47b) 

# Result: 477 rows (samples) Ã— 47 columns (genes)
```

### Step 3.2: 344 samples having mutations in 151 (150) remaining LUAD drivers

```python
# Prepare index
dfMat_LUAD150a = dfindex.copy()
 
# For loop to append columns of 150 driver genes
for i in list(range(len(driv6_150new))): 
    dfMat_LUAD150a[driv6_150new[i]] = np.array(df32[driv6_150new[i]])

# Keep only samples having at lease one mutated genes
dfMat_LUAD150b = dfMat_LUAD150a[dfMat_LUAD150a.sum(axis=1) >= 1]
with pd.option_context('display.max_rows',2): display(dfMat_LUAD150b) 

# Result: 344 rows (samples) Ã— 150 columns (genes)
```

### Step 3.3: 484 samples having mutations in 299 cancer genes from FoundationOne

```python
# Prepare index
dfMat_F1CDx299a = dfindex.copy()
 
# For loop to append columns of 299 cancer genes
for i in list(range(len(driv10_F1CDx299))): 
    dfMat_F1CDx299a[driv10_F1CDx299[i]] = np.array(df32[driv10_F1CDx299[i]])

# Keep only samples having at lease one mutated genes
dfMat_F1CDx299b = dfMat_F1CDx299a[dfMat_F1CDx299a.sum(axis=1) >= 1]
with pd.option_context('display.max_rows',2): display(dfMat_F1CDx299b) 

# Result: 484 rows (samples) Ã— 299 columns (genes)
```

### Step 3.4: 448 samples having mutations in 47 cancer genes from Trueseq

```python
# Prepare index
dfMat_Truseq47a = dfindex.copy()
 
# For loop to append columns of 47 Truseq cancer genes
for i in list(range(len(driv11_Truseq47))): 
    dfMat_Truseq47a[driv11_Truseq47[i]] = np.array(df32[driv11_Truseq47[i]])

# Keep only samples having at lease one mutated genes
dfMat_Truseq47b = dfMat_Truseq47a[dfMat_Truseq47a.sum(axis=1) >= 1]
with pd.option_context('display.max_rows',2): display(dfMat_Truseq47b) 

# Result: 448 rows (samples) Ã— 47 columns (genes)
```

### Step 3.5: 475 samples having mutations in 121 cancer genes from PanCan

```python
# Prepare index
dfMat_Pancan121a = dfindex.copy()
 
# For loop to append columns of 121 PanCan cancer genes
for i in list(range(len(driv12_Pancan121))): 
    dfMat_Pancan121a[driv12_Pancan121[i]] = np.array(df32[driv12_Pancan121[i]])

# Keep only samples having at lease one mutated genes
dfMat_Pancan121b = dfMat_Pancan121a[dfMat_Pancan121a.sum(axis=1) >= 1]
with pd.option_context('display.max_rows',2): display(dfMat_Pancan121b) 

# Result: 475 rows (samples) Ã— 121 columns (genes)
```

## STEP 4: Conclusion and Export Matrix files to NBS Propagation

```python
# Conclusion of remaining samples and genes:
## From total 505 samples and 16057 genes:

# 1) 47 early LUAD drivers:  From 505 --> reduced to 477 samples with 47 genes
# 2) 151 remaining LUAD drivers:  From 505 --> reduced to 344 samples with 150 genes
# 3) FoundationOne_324:  From 505 --> reduced to 484 samples with 299 genes
# 4) Truseq_48:  From 505 --> reduced to 448 samples with 47 genes
# 5) PanCan_127:  From 505 --> reduced to 475 samples with 121 genes
```

```python
# Step 4.1: Reset index and drop 'tumor_stage' before exporting
dfMat_LUAD47b = dfMat_LUAD47b.reset_index()
dfMat_LUAD47c=dfMat_LUAD47b.drop(['tumor_stage'],axis=1)

dfMat_LUAD150b = dfMat_LUAD150b.reset_index()
dfMat_LUAD150c=dfMat_LUAD150b.drop(['tumor_stage'],axis=1)

dfMat_F1CDx299b = dfMat_F1CDx299b.reset_index()
dfMat_F1CDx299c=dfMat_F1CDx299b.drop(['tumor_stage'],axis=1)

dfMat_Truseq47b = dfMat_Truseq47b.reset_index()
dfMat_Truseq47c=dfMat_Truseq47b.drop(['tumor_stage'],axis=1)

dfMat_Pancan121b = dfMat_Pancan121b.reset_index()
dfMat_Pancan121c=dfMat_Pancan121b.drop(['tumor_stage'],axis=1)

dfMat_LUAD47c.head(2)
```

```python
# Step 4.2: Export pivot table as matrix file as an input in pyNBS propogation

dfMat_LUAD47c.to_csv('./output/6.6_OUT_1)EarlyLUAD47_477sam_toNBS_matrix.tsv', sep="\t",index=False)
dfMat_LUAD150c.to_csv('./output/6.6_OUT_2)RemLUAD150_344sam_toNBS_matrix.tsv', sep="\t",index=False)
dfMat_F1CDx299c.to_csv('./output/6.6_OUT_3)FoundationOne299_484sam_toNBS_matrix.tsv', sep="\t",index=False)
dfMat_Truseq47c.to_csv('./output/6.6_OUT_4)Truseq47_448sam_toNBS_matrix.tsv', sep="\t",index=False) 
dfMat_Pancan121c.to_csv('./output/6.6_OUT_5)PanCan121_475sam_toNBS_matrix.tsv', sep="\t",index=False) 
```
