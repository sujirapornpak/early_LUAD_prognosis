# 1c_198 drivers to Fisher's exact test


```python
##STEP 1.1: Import required python libraries (matplotlib, pandas, numpy).

%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 1.2 : Frequency tables of remaining 198 driver genes

```python
# Open cleaned SNV, pivot file:
df1 = pd.read_csv("./output/6.3_LUAD_snv_mc3_OUT_cleaned_cSIF_pivot_16057genes.tsv",sep='\t',header=0) # 505 rows × 16059 columns   

# Make index from multi-columns:
df2 = df1.set_index(['tumor_stage', 'sample_id'])
df2 = df2.sort_index(ascending=True) 
with pd.option_context('display.max_rows',2): display(df2) # 505 rows × 16057 columns   
```

```python
# open list of 198 driver genes file

df3=pd.read_csv("../01_Gene_lists/SuppTable1_All_Drivers_GeneList.tsv",sep='\t',header=0) 
drivlist198 = df3['PooledLUADdrivers_198'].dropna().unique().tolist()
len(drivlist198) # = 198
```

```python
print(drivlist198)
```

```python
# All genes
gene16057 = df2.columns
print(len(gene16057)) # cleaning level C
```

```python
notlistC = []
for i in list(range(len(drivlist198))):
    if drivlist198[i] not in gene16057: 
        notlistC.append(drivlist198[i])
print(notlistC) #only 1 gene (RAD23B) that is not in the list
```

```python
# So, remove 'RAD23B' from the list
drivlist198.remove('RAD23B')
drivlist197 = drivlist198
print(len(drivlist197))
```

```python
# Cut index
df197dri_cSIF = df2.iloc[:,0:0] 
# For loop to append columns of 197 driver genes
for i in list(range(len(drivlist197))):  
    df197dri_cSIF[drivlist197[i]] = np.array(df2[drivlist197[i]])
    
with pd.option_context('display.max_rows',2): display(df197dri_cSIF) # 505 rows × 197 columns
```

```python
# The .loc attribute is the primary access method
# https://pandas.pydata.org/pandas-docs/version/0.13.1/indexing.html

df197dri_cSIF_stg1 = df197dri_cSIF.loc['Stage I':'Stage IB'] # 276 rows × 33 columns
df197dri_cSIF_stg2 = df197dri_cSIF.loc['Stage II':'Stage IIB'] # 121 rows × 33 columns
df197dri_cSIF_stg3 = df197dri_cSIF.loc['Stage IIIA':'Stage IIIB'] # *82 rows (reduced from 83) × 33 columns
df197dri_cSIF_stg4 = df197dri_cSIF.loc['Stage IV'] # 26 rows × 33 columns
    
## Total remaining samples = 276 (stg1) + 121 (stg2) + 82 (stg3) + 26 (stg4) = 505 samples
```

```python
# Compute frequency of 197 driver genes, for 4 separated stages
cSIF33_stg1 = df197dri_cSIF_stg1.sum(axis=0)
cSIF33_stg2 = df197dri_cSIF_stg2.sum(axis=0)
cSIF33_stg3 = df197dri_cSIF_stg3.sum(axis=0)
cSIF33_stg4 = df197dri_cSIF_stg4.sum(axis=0)

df4 = pd.DataFrame(cSIF33_stg1)
df4.rename(columns={0: 'cSIF33_stg1'},inplace=True)
df4['cSIF33_stg1'] = cSIF33_stg1
df4['cSIF33_stg2'] = cSIF33_stg2
df4['cSIF33_stg3'] = cSIF33_stg3
df4['cSIF33_stg4'] = cSIF33_stg4

with pd.option_context('display.max_rows',4): display(df4)
```

```python
df4.to_csv('./output/6.4_LUAD_snv_mc3_OUT_Freq197-4stg_3cleans.tsv', sep="\t",index=True) 
```
